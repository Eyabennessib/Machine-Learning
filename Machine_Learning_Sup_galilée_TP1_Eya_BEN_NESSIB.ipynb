# Exercice 5 : Interprétation des courbes d’apprentissage 
Les courbes d’apprentissage jouent un rôle important. Elles permettent d’une part de vérifier qu’il n’y a pas 
de sur-apprentissage. D’autre part, elles montrent l’évolution des performances du modèle en fonction de la 
quantité des données d’entrainement qu’on lui fournit. Plus la machine dispose de données pour s’entrainer, 
meilleure sera sa performance. Cependant la performance finit par atteindre un plafond donc il est inutile de 
chercher à collecter plus de données.

```python
from sklearn. model_selection import learning_curve 
N, train_score, val_score = learning_curve(model, X_train, Y_train, train_sizes =np.linspace(0.1,  1.0, 10), cv =5)
print(N)
```

N représente le nombre d'exemple d'entrainement utilisés pour générer la courbe d'apprentissage.

```python
plt.plot(N,val_score.mean(axis=1),label ='validation') 
plt.ylabel('score') 
plt.xlabel('train sizes') 
```

**3. Affichez sur le même graphe que la question précédente, les scores moyens obtenus sur le jeu de données d’entrainement**

```python
plt.plot(N,val_score.mean(axis=1),label ='validation', color="darkorange") 
plt.plot(N,train_score.mean(axis=1),label ='training', color="navy") 
plt.ylabel('score') 
plt.xlabel('train sizes') 
```

4. Interprétez les courbes obtenues. Est-il nécessaire de collecter plus de données ? 

D'après la representation graphique des données de train et de validation, on remarque que les  scores sont plus ou moins proportionnels, la metrique de train vaut 1 depuis les premières iterations(epochs) , et la metrique de validation est fluctuante aux premières iterations, pour se stabiliser apres au allentours de 0.96 aux dernières iterations. On peut conclure que le modele    est perforfant, l entrainement est effectué sans overfitting, on n a pas besoin de collecter d'autres données. pour des predictions ulterieures, on doit choisir un checkpoint (ckpt) relatif à l'une des dernieres iterations ( la ou le score des donnees de  train est elevé et le score des données de validation est le plus elevé)
